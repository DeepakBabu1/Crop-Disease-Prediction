```python
import os
from PIL import Image

def resize_images_in_folder(input_folder, output_folder, target_size=(480,640)):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for class_name in os.listdir(input_folder):
        class_path = os.path.join(input_folder, class_name)
        if os.path.isdir(class_path):
            output_class_path = os.path.join(output_folder, class_name)
            if not os.path.exists(output_class_path):
                os.makedirs(output_class_path)

            for image_name in os.listdir(class_path):
                input_image_path = os.path.join(class_path, image_name)
                output_image_path = os.path.join(output_class_path, image_name)

                try:
                    with Image.open(input_image_path) as img:
                        resized_img = img.resize(target_size, Image.Resampling.LANCZOS)
                        resized_img.save(output_image_path)
                        print(f"Resized: {input_image_path} -> {output_image_path}")
                except Exception as e:
                    print(f"Error resizing {input_image_path}: {e}")


input_folder = r"C:\Users\USER\Desktop\Project\Paddy-disease 2.0\Training\Additional_ds" 
output_folder = r"C:\Users\USER\Desktop\Project\Paddy-disease 2.0\Training\Resized_ds"
resize_images_in_folder(input_folder, output_folder, target_size=(480,640))
```
